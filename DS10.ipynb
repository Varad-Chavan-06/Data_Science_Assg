```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = sns.load_dataset('iris')
dataset
```

```python
dataset.info()
```

```python
for i in dataset.columns[:-1]:

    sns.histplot(dataset[i], bins=30)
    plt.title(i)
    plt.show()
```

```python
for i in dataset.columns[:-1]:

    sns.boxplot(dataset[i], color='r')
    plt.title(i)
    plt.show()
```
