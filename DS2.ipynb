```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
student_id = np.arange(1,51)
math_score = np.random.randint(0,100,size=50)
science_score = np.random.randint(0,100,size=50)
english_score = np.random.randint(0,100,size=50)

marks = pd.DataFrame({
    'Student_id':student_id,
    'Math_score':math_score,
    'Science_score':science_score,
    'English_score':english_score
})
print(marks)
```

```python
# Check for missing values
missing_values = marks.isnull().sum()
print("Missing Values:")
print(missing_values)

# Check for inconsistencies (e.g., negative scores, scores above 100)
inconsistencies = (marks < 0) | (marks > 100)
print("\nInconsistencies:")
print(inconsistencies.any())
```

```python
# Check for outliers using box plots
marks[['Math_score', 'English_score', 'Science_score']].boxplot()
```

```python
# Apply logarithmic transformation to Math_Score variable
marks['Math_score_Log'] = np.log(marks['Math_score'])
print(marks)
```
